<h4>Ecuación de Onda</h4>

La siguiente ecuación es la forma general de la ecuación que describe a las ondas en dimensión $1+1$: "la segunda derivada del desplazamiento de la onda $y$ respecto a la coordenada $x$ es igual al inverso de la velocidad de propagación de la onda al cuadrado por la segunda derivada del desplazamiento de esta con respecto al tiempo."

\begin{equation}
    \frac{\partial^{2} y}{\partial x^{2}} - \frac{1}{v^{2}} \cdot (\frac{\partial^{2}y}{\partial t^{2}}) = 0
\end{equation}

Sabemos que la forma general de la solución de ondas viajeras para está ecuación es $y(x, t) = f(x \pm vt)$ pero la forma especifica depende de las condiciones iniciales y de frontera. 

Observando la siguiente imagen:



Podemos expresar las consideraciones anteriores de la siguiente manera:

\begin{equation}
    \frac{\partial^{2} u}{\partial t^{2}} = c \cdot \frac{\partial^{2}u}{\partial x^{2}}, 0 < t, 0 \leq x \leq t
\end{equation}
\begin{equation}
    u(x, 0) = f(x), 0 \leq x \leq t 
\end{equation}
\begin{equation}
    u(0, t) = u(L,t) = 0, t > 0 
\end{equation}
  
La ecuación anterior es una ecuación diferencial parcial (PDE) llamada ecuación de onda y se puede usar para modelar diferentes fenómenos como cuerdas vibrantes y ondas de propagación.

El término constante C tiene dimensiones de $\frac{m}{s}$ y puede interpretarse como la velocidad de onda.

Resulta que el problema anterior tiene la siguiente solución general:

\begin{equation}
    u(x, t) = f(x - ct) + g(x + ct)
\end{equation}

Podemos pensar que cualquier función que tenga el argumento $x - ct$ o $x + ct$ o una combinación de ambos es una solución a la ecuación de onda. Esto significa que podemos modelar muchas olas diferentes. Ademas, la solución general es una combinación de una onda que se desplaza hacia la izquierda y otra hacia la derecha.

Probaremos con una solución particular:

\begin{equation}
    u(x, t) = \frac{1}{2} (sin(x - ct) + sin(x + ct))
\end{equation}

Implementaremos la siguiente solución en Python:

```python
%matplotlib notebook
import numpy as np
from numpy import pi
import matplotlib.pyplot as plt
import matplotlib.animation as animation

#Fondo de pantalla
plt.style.use('dark_background')

fig = plt.figure()
fig.set_dpi(100)
ax1 = fig.add_subplot(1,1,1)

# Velocidad de la Onda
c = int(input("Ingrese la velocidad de la Onda: "))
# c = 1
# Definimos el espacio en que veremos la Onda
x0 = np.linspace(-10,10,10000)

# Tiempo inicial de la Onda
t0 = int(input("Ingrese el tiempo inicial de la Onda: "))
# t0 = 0

# Incremento de Tiempo 
dt = float(input("Ingrese el Incremento del Tiempo (flotante): "))
# dt = 0.05

formula = raw_input("\nIngrese la Solucion Descrita: ")

# Usamos la solución antes descrita
def u(x,t):
    return eval(formula)
# 0.5*(np.sin(x+c*t) + np.sin(x-c*t))

a = [] # Creamos un arreglo vacio, para guardar los valores de la Onda.

for i in range(500):
    value = u(x0,t0)
    t0 = t0 + dt
    a.append(value) # Guardamos en un arreglo los valores de la onda

k = 0
# Animamos nuestra solucion
def animate(i):
    global k
    x = a[k]
    k += 1
    ax1.clear()
    plt.plot(x0,x,color='cyan')
    plt.grid(True)
    plt.ylim([-2,2])
    plt.xlim([-10,10])
    
anim = animation.FuncAnimation(fig,animate,frames=50,interval=20)
anim.save('eqWave.html')
plt.show()
```
